```python
import urllib.request
from bs4 import BeautifulSoup

foxSource = urllib.request.urlopen('https://www.foxnews.com').read()
```

```python
foxSoup = BeautifulSoup(foxSource, 'lxml')
foxArticles = foxSoup.find_all("article")
for foxArticle in foxArticles:
    foxHeader = foxArticle.find("header")
    if foxHeader:
        foxAnchor = foxHeader.find("a")
        print(foxAnchor.string)
```
